Pulls champion key ids from datadragon. Not much to see here

```python
import pandas as pd
from riotwatcher import LolWatcher, ApiError

with open('data/api-key.txt', 'r') as api:
    API_KEY = api.read()
lol_watcher = LolWatcher(API_KEY)

REGION = "kr"
```

```python
# check league's latest version
latest = lol_watcher.data_dragon.versions_for_region(REGION)['n']['champion']
# get some champions static information
static_champ_list = lol_watcher.data_dragon.champions(latest, False, 'en_US')
```

```python
# get champion info and use champion key as index
champ_name_df=pd.DataFrame(static_champ_list["data"]).T[["id","key","name","stats"]]
champ_name_df[champ_name_df["name"]!=champ_name_df.index]
champ_name_df["key"]=champ_name_df["key"].astype(int)
champ_name_df=champ_name_df.reset_index()
champ_name_df=champ_name_df.set_index("key")
champ_name_df
```

```python
# returning the index to column after sorting it and then not saving the index...
# v silly could fix but its fine
key_name = champ_name_df["id"].sort_index().reset_index()
```

```python
key_name.to_csv("data/champion-key-ids.csv", index=False)
```
